Checking the datatypes of each columns, we noticed that the Order Date and Ship Date are not in their appropriate data type.

### Exploratory Data Analysis (EDA):

- Explore the distribution of sales over time.
- Analyze the top-selling products.
- Investigate the relationship between quantity sold and revenue.
- Identify any trends or patterns in customer behavior.

```python
df.groupby('Order Date')['Sales'].sum()
```

```python
sales_over_time = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
```

```python
sales_over_time
```

```python
plt.figure(figsize=(12,6))
sales_over_time.plot(kind='line', color='green', marker='o')
plt.title('Distribution of Sales Over Time')
plt.ylabel('Sales Over Time')
plt.xlabel('Year')

plt.xticks(sales_over_time.index.astype(int))
plt.grid(True)
```

```python
Top_sales_product = df.groupby(df['Product Name'])['Sales'].sum().reset_index()
```

```python
Top_sales_product
```

```python
Top_sales_product.head(10)
```

```python
Top_sales_product = Top_sales_product.sort_values(by='Sales', ascending=False)

top_n = 10
top_product = Top_sales_product.head(top_n)
top_product
```

```python
plt.figure(figsize=(12,6))
top_product.plot(kind='bar', x='Product Name', y='Sales', color='green')
plt.title('Distribution of top_Product')
plt.ylabel('Sales')
plt.xlabel('Product Name')

plt.grid(True)
```

```python
plt.figure(figsize=(12,6))
top_product.plot(kind='hist', x='Product Name', y='Sales', color='green')
plt.title('Distribution of top_Product')
plt.ylabel('Sales')
plt.xlabel('Product Name')

plt.grid(True)
```

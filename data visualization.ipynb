```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
limes = [0.991, 0.930, 0.952, 0.961, 0.937, 0.906, 0.909, 0.974, 0.905, 0.896, 0.7, 0.898, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.plot(years, limes)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.plot(years, limes)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crops Yields in Kanto")
plt.legend(['Apples', 'Oranges', 'Limes']);
```

```python
plt.plot(years, apples, marker='o', c='red')
plt.plot(years, oranges, marker='*', )
plt.plot(years, limes, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crops Yields in Kanto")
plt.legend(['Apples', 'Oranges', 'Limes']);
```

```python
plt.figure(figsize=(14, 8))
plt.plot(years, apples, marker='o', c='red')
plt.plot(years, oranges, marker='*', )
plt.plot(years, limes, marker='x')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crops Yields in Kanto")
plt.legend(['Apples', 'Oranges', 'Limes']);
```

```python
titanic = sns.load_dataset('titanic')

titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic.dropna(subset=['embarked', 'age'])
```

```python
titanic.info()
```

```python
titanic['age'] = titanic['age'].astype(int)
```

```python
titanic.info()
```

```python
grouped = titanic.groupby(['age', 'embarked']).size().unstack(fill_value=0)
```

```python
grouped
```

```python
grouped.plot(kind='line', ls='dashed', figsize=(12, 6))

plt.title('Number of Passenger by Age and Embarkation')
plt.xlabel("Age")
plt.ylabel('Number of Passengers')
plt.legend(title='Embarkation port', labels= ['Cherboug (C)', 'Queenstown (Q)', 'Southampton (S)']);
```

```python
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Petal Dimensions')

sns.scatterplot(x=flowers_df.petal_length, 
                y=flowers_df.petal_width, 
                hue=flowers_df.species,
                s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
                data=flowers_df);
```
